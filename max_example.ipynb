````python
import pandas as pd
import json
import os
````

````python
def read_json_file(filename):
    with open(filename, 'r') as r:
        result = json.loads(r.read())
    r.close() 
    return result
````

### Iterating through our directory and then reading all of the files
* We use ```os.listdir``` to allow us to traverse our directory we pass it

````python
result = pd.DataFrame()
for json_file in os.listdir('data'):
    ###reading json file
    ###https://realpython.com/python-f-strings/ 
    parsed_json = read_json_file(f"data/{json_file}")
    ###creating a dataframe from the json
    new_df = pd.DataFrame.from_dict(parsed_json['GraphImages'])
    ###appending it to the previous dataframe
    result = result.append(new_df)
````

````python
print(result.columns)
result.head()
````

### Selecting the columns we need
* We don't need to do any modifications or querying to our data, so we simply need to select the two columns we need 
* You do this by putting a list of columns in between the brackets, and it will return a new dataframe containing just the columns you selected

````python
timestamp_and_username = result[['taken_at_timestamp', 'username']]
timestamp_and_username
````

### Writing to CSV
* simple

````python
timestamp_and_username.to_csv('timestamp_and_username.csv')
````
